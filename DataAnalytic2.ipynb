```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from pylab import rcParams
rcParams['figure.figsize'] = 12,9
```

```python
corona_dataset_csv = pd.read_csv('covid19_confirmed_global.csv')
corona_dataset_csv.head()
```

```python
df = corona_dataset_csv.drop(["Lat","Long"],axis=1) # axis =1 убираем ненужные столбцы
df.head()
```

```python
corona_dataset_csv.drop(["Long","Lat"],axis=1,inplace=True) # изменяем текущий набор данных
corona_dataset_csv
```

```python
corona_dataset_aggregated = corona_dataset_csv.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregated.head()
```

```python
print("Aggregated shape",corona_dataset_aggregated.shape)
print("Original shape", corona_dataset_csv.shape)
```

```python
corona_dataset_aggregated.loc["China"].plot()
corona_dataset_aggregated.loc["Italy"].plot()
corona_dataset_aggregated.loc["Spain"].plot()
plt.legend()
```

```python
corona_dataset_aggregated.loc["China"][:3].plot() # распространение инфекции за 3 дня в Китае
```

```python
corona_dataset_aggregated.loc["China"].diff().max() # максимум заражений
```

```python
corona_dataset_aggregated.loc["China"].diff().plot() # график показывает уровень заражения по дням, на 11/02/20 приходиться пик заражения
```

```python
corona_dataset_aggregated.loc["China"].diff().max() #. максимальное число зараженных
```

```python
print("Italy с максимальным числом зараженных ", corona_dataset_aggregated.loc["Italy"].diff().max())
print("Spain с максимальным числом зараженных ",corona_dataset_aggregated.loc["Spain"].diff().max())
```

```python
# максимальный уровень инфицирования для всех стран
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated["максимум инфицированных больных"] = max_infection_rates
```

```python
corona_dataset_aggregated.tail()
```

```python
corona_data = pd.DataFrame(corona_dataset_aggregated["максимум инфицированных больных"])
corona_data.head()
```

```python
hapiness_report = pd.read_csv("World_happiness_report_2020.csv")
```

```python
hapiness_report.head()
```

```python
useless_cols = ["Regional indicator","Ladder score","Standard error of ladder score",
               "upperwhisker","lowerwhisker","Generosity","Perceptions of corruption",
               "Ladder score in Dystopia","Explained by: Log GDP per capita",
               
                "Dystopia + residual","Explained by: Perceptions of corruption",
               "Explained by: Generosity","Explained by: Freedom to make life choices",
               "Explained by: Healthy life expectancy","Explained by: Social support"]
```
